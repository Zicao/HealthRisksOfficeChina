```python
import pandas as pd
from bertopic import BERTopic
```

```python
def filter_float_lists(data):
    """
    判断列表中的元素是否为float，并删除长度较短的内容
    data: 包含多个列表的列表
    返回：过滤后的结果
    """
    # 第一步：筛选出所有元素都是float的列表
    float_lists = []
    float_lists= [str(lst) for lst in data]

    if not float_lists:
        return []
    
    # 筛选出长度大于等于平均长度的列表
    filtered_lists = [lst for lst in float_lists if len(lst) >= 10]
    
    return filtered_lists
```

```python
path='.\\Other\\'
filename='111523Cleaned办公室空气噪声空调窗户温度2122232425.csv'
```

```python
df=pd.read_csv(path+filename)
docs=df['Content'].values.tolist()
```

```python
docs=filter_float_lists(docs)
```

```python
topic_model = BERTopic(language="multilingual", calculate_probabilities=True, verbose=True)
topics, probs = topic_model.fit_transform(docs)
```
